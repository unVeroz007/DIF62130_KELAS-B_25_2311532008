```python
import pandas as pd #memanggil library pandas (alat untuk mengolah data) dan menyederhanakan namanya jadi pd agar lebih praktis saat digunakan.
data = {'luas' : [6, 8, 10, 14, 18], #membuat kamus (dictionary) bernama data yang berisi kolom luas dan juga harga dengan isinya masing masing menjadi pasangannya
        'harga' : [7, 9, 13, 17.5, 18]}
df = pd.DataFrame(data) #mengubah kamus data menjadi tabel/dataframe menggunakan pandas. luas, harga) menjadi nama kolom, Angka-angka dalam list menjadi isi baris tabel.
df #menampilkan tabel df yang sudah dibuat.
```

```python
import matplotlib.pyplot as plt #memanggil library matplotlib.pyplot (alat untuk membuat grafik) dan disingkat menjadi plt agar lebih mudah ditulis.
df.plot(kind='scatter', x='luas', y='harga') #membuat scatter plot (titik-titik data) dari DataFrame df, x='luas': Sumbu horizontal (X) menggunakan kolom luas, y='harga': Sumbu vertikal (Y) menggunakan kolom harga.
plt.xlabel('Luas Tanah (m persegi)') #Menambahkan keterangan pada sumbu X dan Y, Menjelaskan bahwa angka-angka di bawah adalah luas tanah dalam meter persegi.
plt.ylabel('Harga Tanah (dollar)') #Menjelaskan bahwa angka-angka di kiri adalah harga dalam dollar.
plt.xlim(0, 25) #Mengatur rentang angka yang ditampilkan pada sumbu. Sumbu X: Dimulai dari 0 hingga 25 (meskipun data luas tertinggi hanya 18 mÂ²).
plt.ylim(0, 25) #Sumbu Y: Dimulai dari 0 hingga 25 (meskipun data harga tertinggi hanya 18 dollar).
plt.grid(True) #Menampilkan garis-garis kotak (grid) di belakang titik data.
plt.show() #Perintah ini menampilkan grafik ke layar
```

```python
import numpy as np #memanggil library NumPy (alat untuk operasi matematika dan array) dan disingkat jadi np

X=np.array(df['luas']) #Independet Variabel (Features), Mengambil kolom luas dari DataFrame df, np.array(...): Mengubah data tersebut menjadi array NumPy (struktur data khusus untuk perhitungan matematis). X: Simpan hasilnya ke variabel X
y=np.array(df['harga']) #Dependent Variabel (Target), df['harga']: Mengambil kolom harga dari DataFrame df, np.array(...): Mengubahnya menjadi array NumPy, y: Simpan hasilnya ke variabel y

print(f'X: {X}') #menampilkan X
print(f'y: {y}') #menampilkan Y
```

```python
X = X.reshape(-1, 1) #Mengubah bentuk (shape) array X dari 1D (deret angka biasa) menjadi 2D (tabel dengan 1 kolom).
#y = y.reshape(-1, 1)

print(f'X shape: {X.shape}') #X.shape menunjukkan bentuk array: 5 baris dan 1 kolom.
print(f'X: {X}') #X menampilkan isi array setelah diubah bentuk.
#print(f'y: {y}')
```

```python
from sklearn.linear_model import LinearRegression #memanggil alat LinearRegression dari library scikit-learn (library untuk machine learning).
model = LinearRegression() #membuat "kotak kosong" bernama model yang siap diisi dengan pola data nanti.
model.fit(X, y) #X (luas tanah) dan y (harga) dimasukkan ke model. Model akan mencari pola hubungan antara X dan y, lalu menghitung persamaan garis terbaik (y = mx + c).
```

```python
X_vis = np.array([0, 25]).reshape(-1, 1) # "np.array([0, 25])" membuat array NumPy dengan dua nilai: 0 dan 25, ".reshape(-1, 1)" Mengubah bentuk array menjadi 2 baris, 1 kolom (matriks 2D) Sebelumnya: [0, 25] (array 1D)
y_vis = model.predict(X_vis) # "model.predict(X_vis)" Model memprediksi hasil berdasarkan nilai X_vis. "y_vis" akan berisi hasil prediksi untuk X_vis = 0 dan X_vis = 25.

print(f'X_vis: {X_vis}') #Menampilkan nilai X_vis
print(f'y_vis: {y_vis}') #Menampilkan hasil prediksi (y_vis)
```

```python
plt.scatter(X, y) #Membuat scatter plot (plot titik-titik) berdasarkan data, X untuk Luas tanah (sumbu X). y untuk Harga tanah (sumbu Y).
plt.plot(X_vis, y_vis, color ='red') # "plt.plot" Menampilkan garis prediksi, "X_vis" untuk Nilai luas tanah yang digunakan untuk prediksi. "y_vis"  Hasil prediksi dari model. "color='red'" untuk Garis prediksi ditampilkan dalam warna merah.
plt.title('Perbandingan luas tanah dengan harga tanah') #Memberikan judul pada grafik: "Perbandingan luas tanah dengan harga tanah".
plt.xlabel('Luas Tanah (m persegi)') # Label sumbu X menunjukkan luas tanah dalam meter persegi.
plt.ylabel('Harga Tanah (dollar)') #Label sumbu Y menunjukkan harga tanah dalam dolar.
#plt.xlim(0, 25) Mengatur batas sumbu X dari 0 sampai 25
#plt.ylim(o, 25) Mengatur batas sumbu Y dari 0 sampai 25.
plt.grid(True) #Menampilkan garis bantu pada grafik untuk mempermudah pembacaan nilai.
plt.show() #Menampilkan grafik ke layar
```

Formula Linear Refression
y = a + bx

y = respon variabel
x = explanatory variable
a = intercept
b = slope

```python
print(f'intercept: {model.intercept_}') #"model.intercept_" adalah Nilai intercept dari model regresi.
print(f'slope: {model.coef_}') #"model.coef_ " adalah Nilai koefisien regresi (kemiringan garis).
```

```python
luas_tanah = np.array([12,15,20, 25]).reshape(-1, 1) #"np.array([12, 15, 20, 25])" Membuat array NumPy berisi nilai 12, 15, 20, dan 25. ".reshape(-1, 1)" untuk Mengubah bentuk array menjadi 2D (matriks dengan 1 kolom dan beberapa baris):
print(luas_tanah.shape) #".shape" untuk Menampilkan ukuran/bentuk array: (4, 1), yang berarti 4 baris, 1 kolom.
luas_tanah #Menampilkan isi variabel "luas_tanah", yaitu array 2D dengan nilai 12, 15, 20, dan 25 dalam satu kolom.
```

```python
harga_prediksi = model.predict(luas_tanah) #"model.predict(luas_tanah)" Menggunakan model machine learning (misalnya regresi linear) untuk memprediksi harga tanah berdasarkan luas_tanah. "harga_prediksi" Menyimpan hasil prediksi harga tanah untuk setiap luas yang diberikan.
harga_prediksi #Menampilkan array yang berisi hasil prediksi harga tanah dari model.
```

```python
for luas, hrg in zip(luas_tanah, harga_prediksi): # "zip(luas_tanah, harga_prediksi)" Menggabungkan dua array (luas_tanah dan harga_prediksi) menjadi pasangan (luas, harga). "for luas, hrg in ..." untuk Mengambil satu per satu pasangan nilai dalam perulangan.
    print(f'Luas: {luas}  Harga prediksi: {hrg}') #Menggunakan f-string untuk mencetak
```

```python
X_train = np.array([6, 8, 10, 14, 18]).reshape(-1, 1) #Membuat array X_train (fitur) yang berisi angka 6, 8, 10, 14, dan 18. ".reshape(-1, 1)" Mengubah array 1D menjadi 2D agar sesuai dengan format yang dibutuhkan model.
y_train = np.array([7, 9, 13, 17.5, 18]) #Membuat array y_train (target/output) yang sesuai dengan X_train.

X_test = np.array([9, 9, 11, 16, 12]).reshape(-1, 1) #X_test â Data fitur uji berisi nilai 9, 9, 11, 16, dan 12 (berbentuk 2D setelah .reshape(-1,1)).
y_test = np.array([11, 8.5, 15, 18, 11]) #y_test â Target/output uji yang sesuai dengan X_test.
```

```python
model=LinearRegression() #Membuat objek model regresi linear dari pustaka scikit-learn. "model" Variabel yang menyimpan model regresi linear yang akan digunakan.
model.fit(X_train, y_train) #Melatih model menggunakan data latih (X_train) dan target yang benar (y_train).
```

Nilai semakin mendekati 1, maka semakin baik

```python
from sklearn.metrics import r2_score #r2_score adalah metrik evaluasi yang digunakan untuk mengukur seberapa baik model regresi linear memprediksi data uji.

y_pred = model.predict(X_test) #"model.predict(X_test)" Menggunakan model yang sudah dilatih untuk memprediksi nilai y (output) berdasarkan X_test (input uji). "y_pred" Menyimpan hasil prediksi model untuk X_test.

r_squared = r2_score(y_test, y_pred) #Menghitung RÂ² Score berdasarkan nilai aktual (y_test) dan nilai prediksi (y_pred).
print(f'R-squared : {r_squared}') #Mencetak nilai RÂ² Score agar kita bisa melihat performa model.
```
